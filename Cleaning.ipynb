```python
data = data.fillna("Other")
data.head()
```

```python
data['Season'].unique()
```

```python
data['Season'] = data['Season'].replace({
    '11-Dec': '11/12',
    'Dec-13': '12/13'
})

data
```

```python
for col in data.columns:
    data[col] = data[col].str.strip().str.upper()
```

```python
data['Competition'].unique()
```

```python
data['Competition'] = data['Competition'].replace({
    'UEFA CHAMPIONS LEAGUE QUALIFYING': 'UEFA CHAMPIONS LEAGUE',
    'CHAMPIONS LEAGUE': 'UEFA CHAMPIONS LEAGUE',
    'TROPH�E DES CHAMPIONS': 'TROPHÉE DES CHAMPIONS',
})

data
```

```python
data.to_csv('data_cleaned.csv', index=False)
```
